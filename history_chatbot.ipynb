```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model = "llama-3.1-8b-instant", groq_api_key=groq_api_key)
```

```python
from langchain_core.messages import HumanMessage, AIMessage
llm.invoke([HumanMessage(content="Hi, I am Aditya Varshney and I am working on Gen AI and Agentic AI")])
```

```python
llm.invoke([HumanMessage(content="What is my name and in which do I work ?")])
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables import RunnableWithMessageHistory
```

```python
store = {}

def get_session_history(session_id:str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

with_message_history = RunnableWithMessageHistory(llm, get_session_history)
```

```python
config1 = {"configurable" : {"session_id" : "user1"}}
config2 = {"configurable" : {"session_id" : "user2"}}
config3 = {"configurable" : {"session_id" : "user3"}}
config4 = {"configurable" : {"session_id" : "user4"}}
```

```python
with_message_history.invoke(
    [
        HumanMessage("Hi, I am Aditya Varshney and I am working in the domain of AI and ML."),
    ],
    config1
)
```

```python
with_message_history.invoke(
    [
        HumanMessage("What is my Name ? and which domain i am working ?"),
    ],
    config1
)
```

```python
with_message_history.invoke(
    [
        HumanMessage("Hi, I am Rahul Gautam and I am working in the domain of GENAI."),
    ],
    config2
)
```
